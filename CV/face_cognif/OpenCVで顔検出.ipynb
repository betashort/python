# 顔検出の手順
## カスケードファイルを指定して検出器を作成
## 対象画像を読み込んで、グレースケールに変換
## 顔検出を実行

```python
#カスケードファイルを指定して、検出器を作成
cascade_file = "haarcascade_frontalface_alt.xml"
cascade = cv2.CascadeClassifier(cascade_file)

#画像の読み込んでグレースケール
img_gray = cv2.cvtColor(img_iq, cv2.COLOR_RGB2GRAY)
#顔認識
face_list = cascade.detectMultiScale(img_gray, minSize = (150, 150))
#結果
if len(face_list) == 0:
    print("失敗")
    quit()
    
for (x,y,w,h) in face_list:
    print("顔の座標 = ", x, y, w, h)
    red = (0, 0, 255)
    cv2.rectangle(img_iq, (x,y), (x+w, y+h), red, thickness=20)
    
#出力
plt.imshow(cv2.cvtColor(img_iq, cv2.COLOR_BGR2RGB))
plt.show()
```

# モザイクをかける

```python
def mosaic(img, rect, size):
    (x1, y1, x2, y2) = rect
    w = x2-x1
    h = y2-y1
    i_rect = img[y1:y2, x1:x2]
    i_small = cv2.resize(i_rect, (size, size))
    i_mos = cv2.resize(i_small, (w, h), interpolation = cv2.INTER_AREA)
    img2 = img.copy()
    img2[y1:y2, x1:x2] = i_mos
    return img2

mos = mosaic(img_iq, (50, 50, 450, 450), 10)
#出力
plt.imshow(cv2.cvtColor(mos, cv2.COLOR_BGR2RGB))
plt.show()
```
