```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.decomposition import NMF
```

```python
data = pd.read_csv("./digit-recognizer/train.csv")
```

```python
data.shape
```

```python
labels = data["label"].iloc[:500]
data = data.drop("label", axis=1).iloc[:500]
```

```python
nmf = NMF(n_components=2)
nmf.fit(data)
```

```python
data_nmf = nmf.transform(data)
```

```python
for i in range(10):
    plt.scatter(
        data_nmf[labels[labels==i][:500].index,0],
        data_nmf[labels[labels==i][:500].index,1],
        label=i)
plt.legend()
plt.show()
```

# PCA

```python
from sklearn.decomposition import PCA
 
pca = PCA(n_components=2)
 
pca.fit(data)
 
data_pca = pca.transform(data)
```

```python
for i in range(10):
    plt.scatter(
        data_pca[labels[labels==i][:500].index,0],
        data_pca[labels[labels==i][:500].index,1],
        label=i)
plt.legend()
plt.show()
```

# t-SNE

```python
from sklearn.manifold import TSNE

data_tsne = TSNE(n_components=2).fit_transform(data)

for i in range(10):
    plt.scatter(
        data_tsne[labels[labels==i][:500].index,0],
        data_tsne[labels[labels==i][:500].index,1],
        label=i)
plt.legend()
plt.show()
```
