```python
%matplotlib inline
import numpy as np
from pylab import *

#毎回同じ乱数になる
np.random.seed(0)

x = np.random.normal(3.0, 1.0, 100)
x_train = x[:80]
x_test = x[80:]

y = np.random.normal(30.0, 2.0, 100)
y_train = y[:80]
y_test = y[80:]
plt.show(scatter(x, y))
plt.show(scatter(x_train, y_train))
plt.show(scatter(x_test, y_test))
```

```python
x = np.array(x)
y = np.array(y)
p4 = np.poly1d(np.polyfit(x, y, 1))
```

```python
xp = np.linspace(0, 7, 100)
plt.scatter(x,y)
plt.plot(xp, p4(xp))
plt.show()
```

```python
slope, intercept = np.polyfit(x, y, 1)
```

```python
print("傾き:{0}\n切片：{1}".format(slope,intercept))
y_true = slope * x + intercept
plt.scatter(x,y)
plt.plot(x, y_true)
plt.show()
```
