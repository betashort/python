```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

```python
dfs = [2, 20, 30]
mk = [',', 'x', '.']
mc = ['red', 'orange', 'black']
x = np.arange(-4, 4, 0.1)
```

```python
for i, df in enumerate(dfs):
    y = st.t.pdf(x, df)
    plt.plot(x, y, color=mc[i], label='df='+str(df), marker=mk[i])
    plt.legend()
```

```python
x_F = np.arange(0.1, 4, 0.1)
dfn = 20
```

```python
for i, dfd in enumerate(dfs):
    y = st.f.pdf(x_F, dfn, dfd)
    plt.plot(x_F, y, color=mc[i], label='df='+str(df), marker=mk[i])
    plt.legend()
```
